```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36'}
url = 'https://exchanges.webmd.com/sexual-conditions-and-stds-exchange/forum/index'
```

```python
r = requests.get(url, headers= headers)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
full_list=[]
for master_page in range(2,10):
  master_page_url="https://exchanges.webmd.com/sexual-conditions-and-stds-exchange/forum/index?pg="+str(master_page)
  headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36'}

  r1 = requests.get(master_page_url, headers= headers)
  soup1 = BeautifulSoup(r1.text, 'html.parser')

  thread_names = soup1.find_all('span',{'class':'thread_fmt'})

  for j in range(len(thread_names)):
    print(j)
    temp_list=[]
    thread_url= soup1.find_all('span',{'class':'thread_fmt'})[j].find('a')['href']
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36'}

    r2 = requests.get(thread_url, headers= headers)
    soup2 = BeautifulSoup(r2.text, 'html.parser')

    #thread_question
    try:
      thread_header= soup2.find('div',{'class':'first_item_title_fmt'}).text

    except:
      thread_header= 'NA'

    #Thread_started_by
    try:
      thread_having_persons= soup2.find_all('div',{'class':'post_hdr_fmt'})[0].text
      thread_having_persons=thread_having_persons.split(' ')
      thread_started_by= thread_having_persons[0].replace('\n','')
      action_performed= thread_having_persons[1]
    except:
      thread_started_by= 'NA'
      action_performed= 'NA'
    #Thread_action performed
    

    #thread_initiated_at
    #if j==0:
    try:
      thread_start_time= soup2.find('div',{'class':'first_posted_fmt'}).find('script').text
      thread_start_time= thread_start_time.replace('document.write(DateDelta(\'', '').replace('\'));', '')
    except:
      thread_start_time='NA'
    #thread_last_reply_at
    try:
      thread_last_reply_at= soup2.find('div',{'class':'firstitem_top_fmt'}).find('script').text
      thread_last_reply_at= thread_last_reply_at.replace('\n\t\t\t\t\t  document.write(DateDelta(\'', '').replace('\'));\n\t\t\t\t', '')
    
    except:
      thread_last_reply_at= 'NA'
    #thread URL
    thread_url_=thread_url

    #thread_content
    try:
      thread_content= soup2.find_all('div',{'class':'post_fmt'})[0].text
    except:
      thread_content='NA'
    temp_list=[thread_header, thread_started_by, action_performed,thread_start_time, thread_last_reply_at,thread_url_,thread_content]
    full_list.append(temp_list)
```

```python
import pandas as pd
df = pd.DataFrame(full_list,columns=("Question","Postby","Action_Performed","Thread_Start_Time", "Last Reply At","Thread Url","Content"))
```

```python
full_final_list=[]
for index, row in df.iterrows():
    temp_list1 =[]
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36'}

    url_each_thread= row['Thread Url']
    print(url_each_thread)

    r_final = requests.get(url_each_thread, headers= headers)
    soup3 = BeautifulSoup(r_final.text, 'html.parser')
    try:
      thread_response_time= soup3.find_all('div',{'class':'posted_fmt'})#.find('script').text
    except:
      thread_response_time=0

    if thread_response_time==0 or len(thread_response_time)==0:
      response_author_name_='NA'
      response_timestamp='NA'
      response_text='NA'
      temp_list1= [row['Question'], row['Postby'], row['Action_Performed'], row['Thread_Start_Time'], row['Last Reply At'], row['Thread Url'], row['Content'],response_author_name_, response_timestamp, response_text]
      full_final_list.append(temp_list1)
    elif thread_response_time!=0:
      print('len(thread_response_time: ',len(thread_response_time))
      
      for k in range(0,len(thread_response_time)):
        print('k=', k)
        ##
        try:
          response_author_name_= soup3.find_all('div',{'class':'post_hdr_fmt'})[k+1].find('a').text
          #response_author_name=response_author_name.split(' ')
          #response_author_name_= response_author_name[0]
        except:
          response_author_name_= soup3.find_all('div',{'class':'post_hdr_fmt'})[k+1].text
          response_author_name_=response_author_name_.split(' ')[0].replace('\n','')

        ##
        ##response_action= response_author_name[1]
        
        ##
        response_timestamp =soup3.find_all('div',{'class':'posted_fmt'})[k].find('script').text
        response_timestamp=response_timestamp.replace('document.write(DateDelta(\'', '').replace('\'));', '')

        ##
        response_text= soup3.find_all('div',{'class':'post_fmt'})[k+1].text

        temp_list1= [row['Question'], row['Postby'], row['Action_Performed'], row['Thread_Start_Time'], row['Last Reply At'], row['Thread Url'], row['Content'],response_author_name_, response_timestamp, response_text]
        full_final_list.append(temp_list1)
```

```python
df_final = pd.DataFrame(full_final_list,columns=("Master_Question","Master_Postby","Master_Action","MasterPost_timestamp","MasterPost_Last Reply_Received_At","Master_Thread Url","MasterPost_Content","response_author_name","response_timestamp","response_text"))
```
